```python
import numpy as np
import pandas as pd
import matplotlib as plt
import math
import itertools
import re
```

```python
#All together now!

#Needs to be documented

# Load in the distance formula
def distance(p0, p1):
    return abs(p0[0] - p1[0]) + abs(p0[1] - p1[1])

pnt_match = 0

distances = []

#imput method

day1 = pd.read_csv(r'C:\Users\Travis\Desktop\Results\f_d1.csv')
day2 = pd.read_csv(r'C:\Users\Travis\Desktop\Results\f_d2.csv')

# Changing the dictionary to a list of tupples to later send them to the distance formula
a_list_day1 = list(zip(day1['Longitude'], day1['Latitude'])) 

# Changing the dictionary to a list of tupples to later send them to the distance formula
a_list_day2 = list(zip(day2['Longitude'], day2['Latitude']))

for a,b in itertools.product(a_list_day1, a_list_day2):
    how_far = distance(a,b)
    distances.append(how_far)
    
    
if not 'plant_name' in day1.columns:
    print('first time matching')
    day1['plant_name'] = 'none'

    day2['plant_name'] = 'none'

    while len(a_list_day2) >= 1 and len(a_list_day1) >=1:

        for a,b in itertools.product(a_list_day1, a_list_day2):
            how_far = distance(a,b)
            #print('Finding new match...')

            if len(distances) != 0:
                if how_far == min(distances):
                    match_1_lat = a[1]
                    match_1_long = a[0]
                    match_2_lat = b[1]
                    match_2_long  = b[0]
                    print('Match found...')

                    for index, row in day1.iterrows():
                        if (day1.Latitude[index] == match_1_lat) & (day1.Longitude[index] == match_1_long) & (day1.plant_name[index] == 'none'):
                            day1.plant_name[index] = str(pnt_match)
                            print('match 1 name:', f'{pnt_match}')


                    for index, row in day2.iterrows():
                        if (day2.Latitude[index] == match_2_lat) & (day2.Longitude[index] == match_2_long) & (day2.plant_name[index] == 'none'):
                            day2.plant_name[index] = str(pnt_match)
                            print('match 2 name:', f'{pnt_match}')
                            pnt_match += 1
                            print('Match Finished')
                            print(f'{len(a_list_day1)}', 'Matches Left...')
                            distances.clear()
                            a_list_day1.remove(a)
                            a_list_day2.remove(b)

                            #Rebuilding the Distance list
                            for a,b in itertools.product(a_list_day1, a_list_day2):
                                how_far = distance(a, b)
                                distances.append(how_far)

else:
    
    day2['plant_name'] = 'none'
    print('new matching pair')
    while len(a_list_day2) >= 1 and len(a_list_day1) >=1:

        for a,b in itertools.product(a_list_day1, a_list_day2):
            how_far = distance(a,b)
            #print('Finding new match...')

            if len(distances) != 0:
                if how_far == min(distances):
                    match_1_lat = a[1]
                    match_1_long = a[0]
                    match_2_lat = b[1]
                    match_2_long  = b[0]
                    print('Match found...')

                    for index, row in day1.iterrows():
                        if (day1.Latitude[index] == match_1_lat) & (day1.Longitude[index] == match_1_long):
                            name = day1.plant_name[index]
                            print('got the name')

                    for index, row in day2.iterrows():

                        if (day2.Latitude[index] == match_2_lat) & (day2.Longitude[index] == match_2_long) & (day2.plant_name[index] == 'none'):
                            day2.plant_name[index] = name
                            pnt_match += 1
                            print('Match Finished')
                            print(f'{len(a_list_day1)}', 'Matches Left...')
                            distances.clear()
                            a_list_day1.remove(a)
                            a_list_day2.remove(b)

                            #Rebuilding the Distance list
                            for a,b in itertools.product(a_list_day1, a_list_day2):
                                how_far = distance(a, b)
                                distances.append(how_far)
print('All Items Matched.')
day1.to_csv(r'C:\Users\Travis\Desktop\Results\day1_matchnosq.csv')
day2.to_csv(r'C:\Users\Travis\Desktop\Results\day2_matchnosq.csv')
```
